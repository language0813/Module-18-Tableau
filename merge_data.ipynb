```python
# Import dependency
import pandas as pd
```

```python
# Read year 2019 csv files downloaded from citibike website with the Pandas library
jan_2019 = pd.read_csv("Raw_Data/JC-201901-citibike-tripdata.csv")
feb_2019 = pd.read_csv("Raw_Data/JC-201902-citibike-tripdata.csv")
mar_2019 = pd.read_csv("Raw_Data/JC-201903-citibike-tripdata.csv")
apr_2019 = pd.read_csv("Raw_Data/JC-201904-citibike-tripdata.csv")
may_2019 = pd.read_csv("Raw_Data/JC-201905-citibike-tripdata.csv")
jun_2019 = pd.read_csv("Raw_Data/JC-201906-citibike-tripdata.csv")
jul_2019 = pd.read_csv("Raw_Data/JC-201907-citibike-tripdata.csv")
aug_2019 = pd.read_csv("Raw_Data/JC-201908-citibike-tripdata.csv")
sep_2019 = pd.read_csv("Raw_Data/JC-201909-citibike-tripdata.csv")
oct_2019 = pd.read_csv("Raw_Data/JC-201910-citibike-tripdata.csv")
nov_2019 = pd.read_csv("Raw_Data/JC-201911-citibike-tripdata.csv")
dec_2019 = pd.read_csv("Raw_Data/JC-201912-citibike-tripdata.csv")
```

```python
# To reduce the number of data files before importing them to Tableau, combine year 2019 monthly data into a dataframe by stacking rows
combined_2019_df = pd.concat([jan_2019, feb_2019, mar_2019, apr_2019, may_2019, jun_2019, jul_2019, aug_2019, sep_2019, oct_2019, nov_2019, dec_2019], ignore_index=True)

# Preview the dataframe
combined_2019_df
```

```python
# Add "age" column for user's age in year 2019 based on "birth year" column
combined_2019_df['age'] = 2019 - combined_2019_df['birth year']
combined_2019_df
```

```python
# Read year 2020 monthly csv files downloaded from citibike website with the Pandas library
jan_2020 = pd.read_csv("Raw_Data/JC-202001-citibike-tripdata.csv")
feb_2020 = pd.read_csv("Raw_Data/JC-202002-citibike-tripdata.csv")
mar_2020 = pd.read_csv("Raw_Data/JC-202003-citibike-tripdata.csv")
apr_2020 = pd.read_csv("Raw_Data/JC-202004-citibike-tripdata.csv")
may_2020 = pd.read_csv("Raw_Data/JC-202005-citibike-tripdata.csv")
jun_2020 = pd.read_csv("Raw_Data/JC-202006-citibike-tripdata.csv")
jul_2020 = pd.read_csv("Raw_Data/JC-202007-citibike-tripdata.csv")
aug_2020 = pd.read_csv("Raw_Data/JC-202008-citibike-tripdata.csv")
sep_2020 = pd.read_csv("Raw_Data/JC-202009-citibike-tripdata.csv")
oct_2020 = pd.read_csv("Raw_Data/JC-202010-citibike-tripdata.csv")
nov_2020 = pd.read_csv("Raw_Data/JC-202011-citibike-tripdata.csv")
dec_2020 = pd.read_csv("Raw_Data/JC-202012-citibike-tripdata.csv")
```

```python
# To reduce the number of data files before importing them to Tableau, combine year 2019 monthly data into a dataframe by stacking rows
combined_2020_df = pd.concat([jan_2020, feb_2020, mar_2020, apr_2020, may_2020, jun_2020, jul_2020, aug_2020, sep_2020, oct_2020, nov_2020, dec_2020], ignore_index=True)

# Preview the dataframe
combined_2020_df
```

```python
# Add "age" column for user's age in year 2020 based on "birth year" column
combined_2020_df['age'] = 2020 - combined_2020_df['birth year']
combined_2020_df
```

```python
# Combine the 2019 and 2020 datasets into one consolidated dataset
combined_2019_2020 = pd.concat([combined_2019_df, combined_2020_df], ignore_index=True)
combined_2019_2020
```

```python
# Check the unique gender categories and value counts of each category
# Zero=unknown; 1=male; 2=female
combined_2019_2020['gender'].value_counts()
```

```python
# Replace the original categories with "unknown", "male", and "female" for readability
combined_2019_2020['gender'] = combined_2019_2020['gender'].replace({1: 'Male', 2: 'Female', 0: 'Unknown'})
combined_2019_2020
```

```python
# Check all the values are replaced successfully
combined_2019_2020['gender'].value_counts()
```

```python
# Double check the combined dataset information
combined_2019_2020.info()
```

```python
# Save the combined dataset into a csv file for further process in Tableau
combined_2019_2020.to_csv("Resource_Tableau/combined_citibike_2019_2020.csv", index=False)
```
